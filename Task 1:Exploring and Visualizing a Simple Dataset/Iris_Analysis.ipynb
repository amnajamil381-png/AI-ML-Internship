Visualizing the dataset
- Scatterplot to find does the petal length increase when the petal width increases?

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(data=iris,
                x='PetalLengthCm',
                y='PetalWidthCm',
                hue='Species')       # 'hue' means "color by this column"

plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

- Histograms to see the distribution, are most flowers clustured around a certain value?

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7))
# This creates a 2x2 grid of 4 charts

features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, feature in enumerate(features):
    row = i // 2      # 0,0,1,1  → which row
    col = i % 2       # 0,1,0,1  → which column

    sns.histplot(data=iris, x=feature, hue='Species',
                 ax=axes[row][col], kde=True)
    # kde=True adds a smooth curve over the bars

    axes[row][col].set_title(f'Distribution of {feature}')

plt.tight_layout()   # prevents charts from overlapping
plt.show()
```

- Box plots to detect outliers

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7))

features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, feature in enumerate(features):
    row = i // 2
    col = i % 2

    sns.boxplot(data=iris, x='Species', y=feature,
                ax=axes[row][col])

    axes[row][col].set_title(f'Box Plot of {feature}')

plt.tight_layout()
plt.show()
```
